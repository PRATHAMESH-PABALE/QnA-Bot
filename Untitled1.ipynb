```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

model = SentenceTransformer('all-MiniLM-l6-v2')

qa_data = {
    'What is AI': 'Artificial Intelligence',
    'What is Cricket': 'Cricket is a emotion',
    'who won champions trophy 2025':'India won the champions trophy',
    'who is the greatest captain of India':'Rohit Sharma is the greatest captain of India'
}

questions = list(qa_data.keys())
question_embeddings = model.encode(questions)
```

```python
import gradio as gr

def get_best_answer(user_query):
    query_embedding = model.encode([user_query])
    scores = cosine_similarity(query_embedding, question_embeddings)[0]
    best_match_index = scores.argmax()
    best_match_score = scores[best_match_index]

    if best_match_score > 0.7:
        return qa_data[questions[best_match_index]]
    else:
        return "I'm sorry, I don't have an answer for that question."
```

```python
iface = gr.Interface(
    fn = get_best_answer,
    inputs = "text",
    outputs = "text",
    title = "AI QnA bot",
    description = "Ask me questions about AI",
)


iface.launch(share=True)
```
